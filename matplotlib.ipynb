# **MATPLOTLIB**

   **X-Axis** input var / independent var / date / time / categorical

   **Y-Axis** output var / dependent var
   

---


(1) **LINE-PLOT**

   **x-axis** continous(time , month , year)

   **y-axis** continous numeric

   **use case** showing trends time , tracking sales etc


---



(2) **SCATTER PLOT**

   **x-axis** continous numeric

   **y-axis** continous numeric

   **use case** showing relationship or co relation ship b/w two var


---



(3) **BAR-CHART**

   **x-axis** categorical data (e.g., product names).

   **y-axis** numeric data (e.g., product sales).

   **use case** compare different categories (comparing sales of different products)


---



(4) **PIE-CHART**

   both axis not work only single variable used

   **use case** showing data in percent wise


---



(5) **HISTOGRAM**

   **x-axis** numerical data (e.g., scores , runs).

   **y-axis** frequency (e.g., count of occurance).

   **use case** shows the distribution of a single numeric var (distribution of exam scores)

```python
battersDF = pd.read_csv('/content/drive/MyDrive/matplotlib-01/batter.csv')
tempBatters = battersDF.head(100).sample(25 , random_state=5)
plt.figure(figsize=(20,10))

plt.scatter(tempBatters['avg'],tempBatters['strike_rate'],s=tempBatters['runs'])

for i in range(tempBatters.shape[0]):
  plt.text(tempBatters['avg'].values[i],tempBatters['strike_rate'].values[i],s=tempBatters['batter'].values[i])

```

**Horizontal and Vertical lines**

```python
plt.figure(figsize = (15,10))
plt.scatter(tempBatters['avg'],tempBatters['strike_rate'],s=tempBatters['runs'])
plt.axhline(130 , color = 'orange')
plt.axhline(140 , color = 'red')
plt.axvline(30 , color = 'green')

for i in range(tempBatters.shape[0]):
  plt.text(tempBatters['avg'].values[i],tempBatters['strike_rate'].values[i],s=tempBatters['batter'].values[i])
```

**3D Surface Plots**

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
xx.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig,ax = plt.subplots(figsize=(7,7))
ax = plt.axes(projection='3d')
p = ax.plot_surface(xx,yy,z , cmap='viridis')
fig.colorbar(p)
plt.show()
```

**Contour Plots**

```python
fig,ax=plt.subplots(figsize=(10,10))
p = ax.contour(xx,yy,z)
fig.colorbar(p)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10,10))
p = ax.contourf(xx,yy,z)
fig.colorbar(p)
plt.show()
```
